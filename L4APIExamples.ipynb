```python
# loading the necessary libraries
import pandas as pd
from randomuser import RandomUser
```

```python
# creating RandomUser object
r = RandomUser()
```

```python
# using this function we get a list of 10 random users
some_list = r.generate_users(10)
some_list
```

```python
# Using get method parametes
name = r.get_full_name()
name
```

```python
# lets say we need 10 users with full name and email address, we can use for loop for the same
for user in some_list:
    print(user.get_full_name(), user.get_email())
```

```python
# generate photos of 10 random users

for user in some_list:
    print(user.get_picture())
```

```python
# To generate a table with information about users, we can write a function containing all desirable parameters.
# The parameters will depend upon tests to be performed
# Then we call the get methods
# then return the pandas dataframe with users

def get_users():
    users = []
    for user in RandomUser.generate_users(10):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),"Email":user.get_email(), "DOB":user.get_dob(),"Picture":user.get_picture()})
    return pd.DataFrame(users)    
```

```python
get_users()
```

```python
df1 = pd.DataFrame(get_users())  
df1.head()
# Hence now we have a pandas dataframe that can be used for testing purposes
```

### FRUTYVICE API

Another common way to use API is through requests library.

```python
import requests
import json
```

```python
# Obtaining the Frutyvice API data
data = requests.get("https://fruityvice.com/api/fruit/all")
```

```python
# Retrieving the results
results = json.loads(data.text)
results
```

```python
# Converting into a pandas dataframe
pd.DataFrame(results)
# Right now the data is in Nested JSON format. Look at the column Nutritions containing multiple sub columns, the data here needs to be flattened or normalize
```

```python
df1 = pd.json_normalize(results)
df1.head()
```

```python
#Get te family and genus of a cherry
cherry = df1.loc[df1['name']=='Cherry']
cherry
```

```python
(cherry.iloc[0]['family']) , (cherry.iloc[0]['genus'])
```

```python
# FINDING OUT HOW MANY CALORIES CONTAINED IN A BANANA
banana = df1.loc[df1['name'] == 'Banana']
banana
```

```python
(banana.iloc[0]['nutritions.calories'])
```
